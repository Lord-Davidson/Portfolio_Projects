```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots that will be created
```

```python
# importing the data from the file

df = pd.read_csv(r'C:\Users\oluba\OneDrive\Documents\Data Analysis Projects\Python Data Correlation\movies dataset.csv')
```

```python
# Viewing the data

df.head()
```

```python
# Finding missing data

df.isnull().sum()
```

```python
df.describe()
```

```python
# Replacing missing values in columns score, votes, budget, gross and runtime with the mean value of each of their columns

score_mean=df['score'].mean()
votes_mean=df['votes'].mean()
budget_mean=df['budget'].mean()
gross_mean=df['gross'].mean()
runtime_mean=df['runtime'].mean()

df['score'].replace(np.nan,score_mean, inplace=True)
df['votes'].replace(np.nan,votes_mean, inplace=True)
df['budget'].replace(np.nan,budget_mean, inplace=True)
df['gross'].replace(np.nan,gross_mean, inplace=True)
df['runtime'].replace(np.nan,runtime_mean, inplace=True)
```

```python
# Replacing missing values in columns rating, released, writer, star, country and company with the most occuring value of each of their columns

rating_counts=df['rating'].value_counts().idxmax()
released_counts=df['released'].value_counts().idxmax()
writer_counts=df['writer'].value_counts().idxmax()
star_counts=df['star'].value_counts().idxmax()
country_counts=df['country'].value_counts().idxmax()
company_counts=df['company'].value_counts().idxmax()

df['rating'].replace(np.nan, rating_counts, inplace=True)
df['released'].replace(np.nan, released_counts, inplace=True)
df['writer'].replace(np.nan, writer_counts, inplace=True)
df['star'].replace(np.nan, star_counts, inplace=True)
df['country'].replace(np.nan, country_counts, inplace=True)
df['company'].replace(np.nan, company_counts, inplace=True)
```

```python
# Checking for missing data
df.isnull().sum()
```

```python
# Data types of the columns

df.dtypes
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Drop column
df.drop(['released'], axis=1, inplace=True)
df
```

```python
# Sort the data by gross in descending order

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Enable scrolling in outputs field of the notebook

# pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df = df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show
```

```python
df.head()
```

```python
# Correlations of the numerical data
# pearson, kendall, spearman

pearson=df.corr(method='pearson', numeric_only=True)
kendall=df.corr(method='kendall', numeric_only=True)
spearman=df.corr(method='spearman', numeric_only=True)

print('pearson \n', pearson)
print('kendall \n', kendall)
print('spearman \n', spearman)

# High correlations between gross and votes
# High correlations between gross and budget only exist in pearson correlation method and not in the others
```

```python
# Plot votes vs gross using seaborn

sns.regplot(x='votes', y='gross', data=df,  scatter_kws={"color": "blue"})
```

```python
# Make a heatmap of the correlations using seaborn

pearson_corr_matrix=df.corr(method='pearson', numeric_only=True)

sns.heatmap(pearson_corr_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
kendall_corr_matrix=df.corr(method='kendall', numeric_only=True)

sns.heatmap(kendall_corr_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
spearman_corr_matrix=df.corr(method='spearman', numeric_only=True)

sns.heatmap(spearman_corr_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# Numerizing the object type columns by generating random categorical numbers
# Add them to the heatmap of the correlations

df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
pearson_corr_matrix=df_numerized.corr(method='pearson')
kendall_corr_matrix=df_numerized.corr(method='kendall')
spearman_corr_matrix=df_numerized.corr(method='spearman')

sns.heatmap(pearson_corr_matrix, annot=True)
plt.title('Correlation Matric for Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
sns.heatmap(kendall_corr_matrix, annot=True)
plt.title('Correlation Matric for Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
sns.heatmap(spearman_corr_matrix, annot=True)
plt.title('Correlation Matric for Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# Unstacking the correlations

pearson_corr_pairs = df_numerized.corr(method='pearson').unstack()
kendall_corr_pairs = df_numerized.corr(method='kendall').unstack()
spearman_corr_pairs = df_numerized.corr(method='spearman').unstack()

print('pearson\n', pearson_corr_pairs, 'kendall\n', kendall_corr_pairs, 'spearman\n', spearman_corr_pairs)
```

```python
# Checking for correlations higher than 0.5

pearson_high_corr = pearson_corr_pairs[(pearson_corr_pairs) > 0.5]
kendall_high_corr = kendall_corr_pairs[(kendall_corr_pairs) > 0.5]
spearman_high_corr = spearman_corr_pairs[(spearman_corr_pairs) > 0.5]

print('pearson\n', pearson_high_corr, '\n', 'kendall\n', kendall_high_corr, '\n', 'spearman\n', spearman_high_corr)
```
